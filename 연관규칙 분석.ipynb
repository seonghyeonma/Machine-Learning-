```python
import pandas as pd 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#리스트의 리스트 형태로 트랩잭션 구성 
dataset = [['빵','우유','씨리얼'],
           ['빵','기저귀','맥주','계란'],
           ['우유','기저귀','맥주','콜라'],
           ['빵','우유','기저귀','맥주',],
           ['빵','우유','기저귀','콜라','삼겹살'],
           ['우유','계란','씨리얼'],
           ['우유','기저귀','콜라','맥주'],
           ['삼겹살','콜라','맥주'],
           ['빵','우유','기저귀','맥주','계란'],
           ['우유','빵'],
           ['우유','씨리얼']]
```

```python
#TransactionEncoder는 리스트의 리스트로 구성된 트랜잭션 데이터를 numpy 배열로 인코딩 하기 위함 
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
te_ary
```

```python
te.columns_ #array 위에 붙어 있을 칼럼이 된다, 해당하는 칼럼에 대해서는 T이거 나머지는 F
```

```python
df = pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#지지도 확인
frequent_itemsets = apriori(df,min_support = 0.3, use_colnames = True) #지지도의 최저값 지정 min_support
frequent_itemsets
#지지도와 아이템 셋을 보여준다 0.5 이상인 값듦만 보여준다 
```

```python
#지지도 내림차순으로 정렬
frequent_itemsets.sort_values('support',ascending=False)
```

```python
#신뢰도 확인 설정안하면 0.8부터 보여줌 
association_rules(frequent_itemsets, metric="confidence",min_threshold=0.3)
```

```python
#향상도 확인 1이상이여야지 의미가 있다
association_rules(frequent_itemsets,metric="lift",min_threshold=1)
```

```python
df_assoc = association_rules(frequent_itemsets,metric="lift",min_threshold=1.2)
df_assoc.sort_values('lift',ascending=False)
```
