## 타이타닉 원핫 인코더

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import OneHotEncoder

#타이타닉 데이터에서 탑승항구 변수를 one hot encoding 해보자 
df= pd.read_csv("data/titanic.csv",index_col=["PassengerId"])
print(df.shape)
df.head()
```

```python
#불필요한 변수 제거 
df = df.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
df.isnull().sum()
```

```python
#Agedㅢ 결측치는 중간값으로 Embarked 결측치는 제거 
df.Age.fillna(df.Age.median(),inplace=True)
df = df.dropna()
df
```

```python
#결측치 제거 이후 index를 재조정 - Passenger ID 제거 
df=df.reset_index(drop=True)
```

```python
df
```

```python
#OneHotEncoder를 encoder라는 변수에 저장 
encoder = OneHotEncoder()
```

```python
#encoder 를 사용하여 Embarked 항목 값으로 학습 
encoder.fit(df[['Embarked']])
```

```python
#encoderd' d의 transform()을 사용해 Embarked변수를 변환해 onehot이라는 변수에 저장
onehot = encoder.transform(df[['Embarked']])
```

```python
#encoder하는 변수를 array 형태로 변환
onehot = onehot.toarray()
```

```python
onehot
```

```python
#encoder하는 변수를 DataFrame 형태로 변환
onehot = pd.DataFrame(onehot)
```

```python
onehot
```

```python
#참고) encoder의 'get_feature_names()'를 사용하면 column 이름을 자동 생성할 수 도 있음 
encoder.get_feature_names()
```

```python
#칼럼 이름을 직접 생성 
onehot.columns = ['C','Q','S']
```

```python
#인코딩 결과 확인
#결측치가 들어가면 인코딩이 안된다 
#결측치 제거 필수 
onehot
```

```python
#기존 df에 인코딩 결과를 병합 
onehot = pd.concat([df,onehot],axis=1)
```

```python
onehot
```

```python
#인코딩 전 'embarked'변수 제거 후 df에 업데이트 
df = onehot.drop('Embarked',axis=1)
df
```
